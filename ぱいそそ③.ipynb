#リストに複数のデータを記憶する

###リストの書き方

リストを作るには***全体を書くカッコで囲み、値をカンマで区切って並べる***。

リスト内の個々の値を***「要素」***という。要素は数値でも文字列でも何でも構わない。

```python
変数 = [値a, 値b, 値c]
#リスト[値a, 値b, 値c]を変数に入れろ
```
リスト内の要素を利用するときは、***変数名の後にカッコで囲んで数値を書く***。この数値を***「インデックス(添字)」***という。

インデックスは、０から開始され、最初の要素が要素０に当たる。

```python
変数 = [値a, 値b, 値c]
#リスト[値a, 値b, 値c]を変数に入れろ
print(変数[0])
#変数の要素０（すなわち値a）を表示せよ
```

````python
#リストを作って利用する

wdays = ['月', '火', '水', '木', '金'] #リスト[文字列「月」, 文字列「火」, 文字列「水」, 文字列「木」, 文字列「金」]を変数wdaysに入れろ

print(wdays[1]) #変数wdaysの要素１を表示しろ
````

````python
#リストの要素を書き換える

wdays = ['月', '火', '水', '木', '金'] #リスト[文字列「月」, 文字列「火」, 文字列「水」, 文字列「木」, 文字列「金」]を変数wdaysに入れろ

wdays[1] = '炎' #変数wdaysの要素１に文字列「炎」を入れろ

print(wdays) #変数wdaysの要素１を表示しろ
````

リストの要素は個別に書き換えが可能。

また、インデックスを指定しない場合、リストがそのまま出力される。

#リストの内容を繰り返し文を使って表示する

***for文***は「for 変数 in リスト:」で１セットである。リストの要素が順番に変数に入り、***リストの要素の数だけインデントされた文が繰り返される***。

```python
for 変数 in リスト:
  繰り返したい処理
```

```yaml
リスト内の要素を変数に順次入れる間、以下を繰り返せ
　　繰り返したい処理
```

````python
#月曜日から金曜日まで表示する

wdays = ['月', '火', '水', '木', '金'] #リスト[文字列「月」, 文字列「火」, 文字列「水」, 文字列「木」, 文字列「金」]を変数wdaysに入れろ

for day in wdays: #変数wdaysの要素を変数dayに順次入れる間、以下を繰り返せ
  print(day, '曜日') #変数dayと文字列「曜日」を表示しろ
````

#リストの一部だけを取り出す

「リスト［開始インデックス:終了インデックス］」の形で書くと、***開始インデックスから終了インデックス直前***までの要素が取り出される。これをリストの***スライス***という。

したがって、***終了インデックスに４と指定した場合、その直前の要素３までが取り出される***。

````python
#for文とスライスを組み合わせる

wdays = ['月', '火', '水', '木', '金'] #リスト[文字列「月」, 文字列「火」, 文字列「水」, 文字列「木」, 文字列「金」]を変数wdaysに入れろ

for day in wdays[1:4]: #変数wdaysの要素1〜要素４直前までを変数dayに順次入れる間、以下を繰り返せ
  print(day, '曜日') #変数dayと文字列「曜日」を表示しろ
````

豆知識
```python
wdays[2:] #要素２から末尾まで
wdays[:4] #戦闘から要素３(インデックス４直前)まで
```

#仕事を１０回繰り返す

回数が決まった繰り返しをしたい場合は、リストの代わりに***range関数***を書く。range関数に１つの引数を渡すと***0〜終了値-1までの整数の並び***が作り出される。「range(5)」であれば「0、1、2、3、4」となる。

````python
#同じメッセージを10回表示する

for cnt in range(10): #0〜数値10直前の範囲内の整数を変数cntに順次入れる間、以下を繰り返せ
  print('ハロー！') #文字列「ハロー！」を表示しろ
````

「10を指定したから10回繰り返される」のではなく、***「10個の数値があるから10回繰り返される」***

````python
#メッセージの中に回数を入れる

for cnt in range(10): #0〜数値10直前の範囲内の整数を変数cntに順次入れる間、以下を繰り返せ
  print(cnt, '回目のハロー！') #変数cntと文字列「回目のハロー！」を表示しろ
````

#開始値を指定して繰り返す

range関数に***2つの引数***を指定した場合、***1つ目が開始値・2つ目が終了値***になる。

終了直前までというルールは変わらないので、「1，2，3，4，5」という整数を作りたい場合は「range(1,6)」とかく必要がある。

````python
#１〜１０の整数で繰り返す

for cnt in range(1, 11): #0〜数値10直前の範囲内の整数を変数cntに順次入れる間、以下を繰り返せ
  print(cnt, '回目のハロー！') #変数cntと文字列「回目のハロー！」を表示しろ
````

#繰り返し文を2つ組み合わせて九九の表を作る

for分の中にfor文を書くと多重ループになる。

````python
#九九の計算をしてみよう

for cnt1 in range(1, 10): #0〜数値10直前の範囲内の整数を変数cnt1に順次入れる間、以下を繰り返せ
  for cnt2 in range(1, 10): #0〜数値10直前の範囲内の整数を変数cnt2に順次入れる間、以下を繰り返せ
    print(cnt1 * cnt2) #変数cnt1と変数cnt２を掛けた結果を表示しろ
````

````python
#九九らしく表示をする

for cnt1 in range(1, 10): #0〜数値10直前の範囲内の整数を変数cnt1に順次入れる間、以下を繰り返せ
  for cnt2 in range(1, 10): #0〜数値10直前の範囲内の整数を変数cnt2に順次入れる間、以下を繰り返せ
    print(cnt1, '✕', cnt2, '=', cnt1 * cnt2) #変数cnt１と文字列「✕」と変数cnt２と文字列「＝」と、変数cnt1と変数cnt２を掛けた結果を表示しろ　
````

#条件式を使って繰り返す

while文は、条件を満たす繰り返しをする文である。whileの後には、TrueやFalseを返す四季や関数などを書く。そのため、同じ繰り返し文のfor文よりもif文に書き方が似通っている。

```python
while 継続条件: # 継続条件が真である限り以下を繰り返せ
  繰り返したい処理
```



````python
#財布がからになるまで繰り返す

wallet = 50000 #数値５００００を変数walletに入れろ

while wallet >= 0: #「変数walletは０以上である」が真である限り以下を繰り返せ
  print(wallet) #変数walletを表示しろ
  wallet = wallet - 5080 #変数walletから数値5080を引いた数を変数walletに入れろ
````

pythonにおける「=」は代入文の記号なので、「wallet = wallet -5080」は成立する。

累算代入文を使って式を省略できる

```python
a = a + 10　 → 　a += 10

a = a - 10　 → 　a -= 10

a = a * 10　 → 　a *= 10

a = a / 10　 → 　a /= 10
```



繰り返し文から脱出したい場合はBreak、戦闘に戻って継続したい場合はContinueを使用する

#総当たり戦の表を作ろう

総当りは、「Aチーム対Bチーム」「Aチーム対Cチーム」のような組み合わせを作ること

単純に考えると、九九と同じような多重ループで作れるはず…？

````python
#単純にすべての組み合わせを並べる

teams = ['A', 'B', 'C', 'D', 'E'] #リスト「A、B、C、D、E」を変数teamsに入れろ

for t1 in teams: #変数teams内の要素を変数t1に順次入れる間、以下を繰り返せ
  for t2 in teams: #変数teams内の要素を変数t2に順次入れる間、以下を繰り返せ
    print(t1, 'vs', t2) #変数t１と文字列「vs」と変数t２を表示しろ
````

````python
#同じチーム同士の対戦を除く

#等しくないことを判定する「！=」演算子を使う

teams = ['A', 'B', 'C', 'D', 'E'] #リスト「A、B、C、D、E」を変数teamsに入れろ

for t1 in teams: #変数teams内の要素を変数t1に順次入れる間、以下を繰り返せ
  for t2 in teams: #変数teams内の要素を変数t2に順次入れる間、以下を繰り返せ
    if t1 != t2: #「変数t１と変数t２が等しくない」が真であるならば以下を実行せよ
      print(t1, 'vs', t2) #変数t１と文字列「vs」と変数t２を表示しろ
````

総当たり戦は、上のコードの半分に当たるので、被りを省いた残り半分だけを表示するプログラムを作れば良い

````python
#総当り完成形

teams = ['A', 'B', 'C', 'D', 'E'] #リスト「A、B、C、D、E」を変数teamsに入れろ

opps = ['A', 'B', 'C', 'D', 'E'] #リスト「A、B、C、D、E」を変数oppsに入れろ

for t1 in teams: #変数teams内の要素を変数t1に順次入れる間、以下を繰り返せ
  opps.remove(t1)
  for t2 in opps: #変数teams内の要素を変数t2に順次入れる間、以下を繰り返せ
    print(t1, 'vs', t2) #変数t１と文字列「vs」と変数t２を表示しろ
````
