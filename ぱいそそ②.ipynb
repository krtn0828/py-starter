#条件分岐ってどんなもの？

***「条件を満たすときだけ実行したいコード」***がある場合、***「条件分岐」***を使用する。

プログラムの流れが分かれるので、**「分岐」**と呼ばれる。

分岐が多くなると流れがわかりづらくなるため、***「フローチャート（流れ図）」***を書いて整理する。

**TrueとFalse**

***True(真)は条件を満たした状態***を表し、***False(偽)は条件を満たしていない状態***を表す。

このような真偽値は文字列や数値と同じ値の一種で、「プール値」と呼ばれる。

#入力されたものが数字かどうか調べる

isdigitメソッドは文字列用のメソッドで、文字列の中身が０〜９の場合はTrue、そうでない場合はFalseを返す。

isdigitは数字かどうかのみを判定するため「−」や「.」が含まれる場合はFalseを返す。

````python
#isdigitメソッドを使ってみる

text = input('入力せよ：　') #文字列「入力せよ：　」を表示してユーザーに入力させ、結果を変数textに入れろ

print(text.isdigit()) #変数textは数字のみか表示しろ
````

#数字が入力されたら計算する

if文は条件分岐の基本となる文である。ifの後に書いた式がTrue(真)であれば、字下げしている部分へ進む。

```python
if 式　:
  実行する処理
```
↓訳
```
もしも「式の結果」が真なら以下を実行せよ
  実行する処理
```




````python
#文字列が数字だったらメッセージを表示する

text = input('入力せよ：　') #文字列「入力せよ：　」を表示してユーザーに入力させ、結果を変数textに入れろ

if text.isdigit(): #もしも「変数textは数字のみ」が真ならば以下を実行せよ
  print('数字です') #文字列「数字です」を表示しろ
````

````python
#数字のときは計算する

text = input('入力せよ：　') #文字列「入力せよ：　」を表示してユーザーに入力させ、結果を変数textに入れろ

if text.isdigit(): #もしも「変数textは数字のみ」が真ならば以下を実行せよ
  print(100 - int(text)) #数値１００から変数textを整数化したものを引いた結果を表示しろ
````

#数字が入力されていない時に警告する

Falseのときにも何かをしたい場合は、if文の後に***else節***を追加する。

```python
if 式:
  実行する処理１
else:
  実行する処理２
```
↓訳

```yaml
もしも「式の結果」が真なら以下を実行せよ
  実行する処理１
そうでなければ以下を実行せよ
  実行する処理２
```





````python
#else節を追加してみる

text = input('数字を入力せよ：　') #文字列「数字を入力せよ」を表示してユーザーに入力させ、結果を変数textに入れろ

if text.isdigit(): #もしも「変数textは数字のみ」が真ならば以下を実行せよ
  print(100 - int(text)) #数値１００から変数textを整数化したものを引いた結果を表示しろ
else: #そうでなければ以下を実行せよ
  print('数字ではない') #文字列「数字ではない」を表示しろ
````

#比較演算子で大小を判定する

年齢層の判定とは「20歳未満なら未成年」「20歳以上なら成人」
というように与えられた数値が基準値より大きいか小さいかで判定される。

このような式を書くには、***比較演算子***を使う

| 演算子 | 読み方                 | 　例       | 
| :----- | :--------------------- | :--------- | 
| <      | 左辺は右辺より小さい   | a < b      | 
| <=     | 左辺は右辺以下         | a <= b     | 
| >      | 左辺は右辺より大きい   | a > b      | 
| >=     | 左辺は右辺以上         | a >= b     | 
| ==     | 左辺と右辺は等しい     | a == b     | 
| !=     | 左辺と右辺は等しくない | a != b     | 
| in     | 左辺は右辺に含まれる   | a in b     | 
| not in | 左辺は右辺に含まれない | a not in b | 

````python
#比較する式の結果を見る①

print(4 < 5) #「数値４はすうち５より小さい」の結果を表示しろ
````

````python
#比較する式の結果を見る②

print(4 > 5) #「数値４は数値５より大きい」の結果を表示しろ
````

````python
#if文と比較する式を組み合わせる

text = input('年齢は？') #文字列「年齢は？」を表示してユーザーに入力させ、結果を変数textに入れろ

age = int(text) #変数textを整数化したものを変数ageに入れろ

if age < 20: #もしも「変数ageが２０以下」が真ならば以下を実行せよ
  print('未成年') #文字列「未成年」を表示しろ
````

#3段階以上に分岐させる

if文に***elif節***を追加すると、if文に複数の条件をもたせることができる。

「そうではなく『〜〜』が真であるならば」と考えると良い。

```python
if 式1 :
  実行する処理1
elif 式2 : #書き方はif節と同じ
  実行する処理2
else: #else節はすべての条件を満たさかなかった時に実行される
  実行する処理3
```



````python
#elif節を使って書いてみる

text = input('年齢は？') #文字列「年齢は？」を表示してユーザーに入力させ、結果を変数textに入れろ

age = int(text) #変数textを整数化したものを変数ageに入れろ

if age < 20: #もしも「年齢は20歳未満」が真であるならば以下を実行せよ
  print('未成年') #文字列「未成年」を表示しろ

elif age < 65: #そうではなく「年齢は65歳未満」が真であるならば以下を実行せよ
  print('成人') #文字列「成人」を表示しろ

else: #そうでなければ以下を実行せよ
  print('高齢者') #文字列「高齢者」を表示せよ
````

if節とelif節を用いたプログラムでは、**上から順に**式が真であるかどうか判定される。

````python
#elif節をif節に変えるとどうなる？

text = input('年齢は？') #文字列「年齢は？」を表示してユーザーに入力させ、結果を変数textに入れろ

age = int(text) #変数textを整数化したものを変数ageに入れろ

if age < 20: #もしも「年齢は20歳未満」が真であるならば以下を実行せよ
  print('未成年') #文字列「未成年」を表示しろ

if age < 65: #もしも「年齢は65歳未満」が真であるならば以下を実行せよ
  print('成人') #文字列「成人」を表示しろ

else: #そうでなければ以下を実行せよ
  print('高齢者') #文字列「高齢者」を表示せよ
````

if節だけで構成すると、**条件に対して真のブロックはすべて実行される**。

#条件分岐の中に条件分岐を書く

数字のみという条件がTrueのときのみに年齢層の判定をしたい場合、「数字のみ」をチェックするif文のブロック内に年齢を判定するif文を書く。

````python
#2つのif文を組み合わせる

text = input('年齢は？') #文字列「年齢は？」を表示してユーザーに入力させ、変数textに入れろ

if text.isdigit(): #もしも「変数textは数字のみ」が真であるならば以下を実行せよ
  age = int(text) #変数textを数値化したものを変数ageに入れろ
  
  if age < 20: #もしも「変数ageが20未満」が真ならば以下を実行せよ
    print('未成年') #文字列「未成年」を表示しろ
````

#複数の比較式を組み合わせる

***プール演算子***はプール値（TrueかFalse）を受け取って結果を返す演算子で、***and、or、not***の3種類がある。

***and演算子***は***左右の値が両方ともTrueのときだけTrueを返す***。→かつ

````python
#義務教育の対象かチェックする

text = input('年齢は？') #文字列「年齢は？」を表示してユーザーに入力させ、結果を変数textに入れろ

age = int(text) #変数textを整数化したものを変数ageに入れろ

if age >= 6 and age <= 15: #もしも「変数ageが6以上、かつ変数ageが15以下」が真であるならば以下を実行せよ
  print('義務教育の対象') #文字列「義務教育の対象」を表示せよ
````

***or演算子***は、***左右のどちらか一方でもTrueのときにTrueを返す***。→または


````python
#幼児と高齢者だけ対象にする

text = input('年齢は？') #文字列「年齢は？」を表示してユーザーに入力させ、結果を変数textに入れろ

age = int(text) #変数textを整数化したものを変数ageに入れろ

if age >= 5 or age <= 65: #もしも「変数ageが5以下、または変数ageが６５以上」が真であるならば以下を実行せよ
  print('幼児もしくは高齢者') #文字列「幼児もしくは高齢者」を表示せよ
````

***not演算子***は、***直後にあるTrueとFalseを逆転させる***。→ではない

````python
#not演算子を使ってFalseのときだけ実行する

text = input('年齢は？') #文字列「年齢は？」を表示してユーザーに入力させ、結果を変数textに入れろ

if not text.isdigit(): #もしも「変数textが数字のみである」が真ではないならば以下を実行せよ
  print('数値を入力して') #文字列「数値を入力して」を表示せよ
````

#年齢層を分析するプログラムを作ってみよう

プログラムの仕様を最初に整理する

*   ユーザーに年齢を入力させる
*   入力された文字列が数字のときだけ年齢層の判定を行う
*   年齢に応じて「未成年」「成人」「高齢者」の3つの結果を表示する
*   未成年のうち、義務教育期間の場合は「未成年（義務教育期間）」と表示する


````python
#数字のみ判定のブロック内に3段階の判定を書く

text = input('年齢は？') #文字列「年齢は？」を表示してユーザーに入力させ、変数textに入れろ

if text.isdigit(): #もしも「変数textは数字のみ」が真であるならば以下を実行せよ
  age = int(text) #変数textを数値化したものを変数ageに入れろ
  
  if age < 20: #もしも「変数ageが20未満」が真ならば以下を実行せよ
    print('未成年') #文字列「未成年」を表示しろ

  elif age < 65: #そうではなく「年齢は65歳未満」が真であるならば以下を実行せよ
    print('成人') #文字列「成人」を表示しろ

  else: #そうでなければ以下を実行せよ
    print('高齢者') #文字列「高齢者」を表示せよ
````

if文において、空の文字列はFalseと判定される。

````python
#義務教育期間の判定を追加する

text = input('年齢は？') #文字列「年齢は？」を表示してユーザーに入力させ、変数textに入れろ

if text.isdigit(): #もしも「変数textは数字のみ」が真であるならば以下を実行せよ
  age = int(text) #変数textを数値化したものを変数ageに入れろ
  
  if age < 20: #もしも「変数ageが20未満」が真ならば以下を実行せよ
    
    if age >= 6 and age <= 15: #もしも「変数ageが6以上、かつ変数ageが15以下」が真であるならば以下を実行せよ
      print('義務教育の対象') #文字列「義務教育の対象」を表示せよ

    else: #そうでなければ以下を実行せよ
      print('未成年') #文字列「未成年」を表示しろ

  elif age < 65: #そうではなく「年齢は65歳未満」が真であるならば以下を実行せよ
    print('成人') #文字列「成人」を表示しろ

  else: #そうでなければ以下を実行せよ
    print('高齢者') #文字列「高齢者」を表示せよ
  
````
